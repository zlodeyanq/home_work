## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
all_gender = data['Gender'].count()
males = data['Gender'][data['Gender'] == 'Male'].count()
females = data['Gender'][data['Gender'] == 'Female'].count()
other = data['Gender'][data['Gender'] == 'Other / Non-Disclosed'].count()


def percentage_gender(x):
    percentage = x / all_gender * 100
    return(f'Percentage of total {int(percentage)}%')


males = percentage_gender(males)
females = percentage_gender(females)
other = percentage_gender(other)

gender_df = pd.DataFrame({'stasts': [males, females, other]})
gender_df



    
    

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Create a groupby object
gender_group = data.groupby('Gender')

#Make calculations
avr_gender = gender_group.mean()
sum_gender = gender_group.sum()
count_gender = gender_group.count()

#Display summary in DataFrame
mega_df = pd.DataFrame({
    'Average price': avr_gender['Price'],
    'Total purchased': sum_gender['Price'],
    'Number of transactions': count_gender['Price']
})

mega_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create brackets for the bins
ages = [1,10,20,40]

#Create bins
bins_categories = ['baby','teenager', 'adult']

#Cut bins
data['Age Categories'] = pd.cut(data['Age'], ages, labels = bins_categories)

#Calculate everything
all_ages = data['Age Categories'].count()
baby = data['Age Categories'][data['Age Categories'] == 'baby'].count()
teenager = data['Age Categories'][data['Age Categories'] == 'teenager'].count()
adult = data['Age Categories'][data['Age Categories'] == 'adult'].count()


def percentage_gender(x):
    percentage = x / all_ages * 100
    return(f'{int(percentage)}%')
    


baby_p = percentage_gender(baby)
teenager_p = percentage_gender(teenager)
adult_p = percentage_gender(adult)


summary_age_category = pd.DataFrame({'Category': ['Baby','Teenager','Adult'],
                        'Number': [baby,teenager, adult],
                        'Percentage':[baby_p, teenager_p, adult_p]
})

summary_age_category
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Upload fresh data 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
data = pd.read_csv(file_to_load)

#Create brackets for the bins
ages = [1,10,20,40]

#Create bins
bins_categories = ['baby','teenager', 'adult']

#Cut bins
data['Age Categories'] = pd.cut(data['Age'], ages, labels = bins_categories)


#Group by Age Categories
age_group = data.groupby('Age Categories')

#Make calculations
avr_age = age_group.mean()
sum_age = age_group.sum()
count_age = age_group.count()

#Display summary in DataFrame
mega_df = pd.DataFrame({
    'Average price': avr_age['Price'],
    'Total purchased': sum_age['Price'],
    'Number of transactions': count_age['Price']
})

mega_df


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_sum = data.groupby(["Item ID", 'Item Name']).sum()

item_mean = data.groupby(["Item ID", 'Item Name']).mean()

item_count = data.groupby(["Item ID", 'Item Name']).count()

total_item = pd.DataFrame({"Items Sum": item_sum['Price'],
                          "Item Average": item_mean['Price'],
                          "Quanitity of items": item_count['Price']})

total_item.head()



```
